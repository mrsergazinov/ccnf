```python
import matplotlib.pylab as plt
import os
import argparse
from config.setup import Config
from model.solver import *
```

```python
# load config file
conf = Config(conf_file_path='./config/electricity.yaml', seed=None, device='cuda:1', exp_name='electricity',
        exp_log_path='./log/electricity/06.28.2022_10.48.00')
```

```python
# load solver
solver = CCNF(conf)
```

```python
test_loader, data_formatter, model, loss = solver.evaluate()
```

```python
device = 'cuda:1'
for i, (x, y, id) in enumerate(test_loader):
    x, y = x.to(device), y.to(device)
    dummy_y = torch.zeros_like(y)
    _, _, _, y_pred = model.g(x, dummy_y)

    #TODO: model should predict quantiles, quantile predictions should be attached along last axis
    # below is temp. fix
    y_pred = y_pred.unsqueeze(-1)

    # compute loss
    # loss, _ = loss(y_pred, y)
    # smape.append(smape_loss(y_pred.squeeze().detach().cpu().numpy(), y.detach().cpu().numpy()))

    # unnormalize tensor
    target = unnormalize_tensor(data_formatter, y, id[0][0])
    p50_forecast = unnormalize_tensor(data_formatter, y_pred.squeeze(), id[0][0])
    
    break
    # compute risk
    p50_risk.append(loss.numpy_normalised_quantile_loss(p50_forecast, target, 0.5))
```
